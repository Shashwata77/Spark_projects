## IMPORTING BASIC LIBRARIES

```python
import matplotlib as mp
import seaborn as sb
import pandas as pd 
import numpy as np
from numpy import math
from matplotlib import pyplot as plt
import seaborn as sb
```

## LOADING THE DATASET

```python
df=pd.read_csv("iris1.csv")
df.head()
```

## TRYING TO PLOT THE COUNT OF EVERY CLASS

```python
sb.countplot(x="Species",data=df,palette="CMRmap")
plt.show()
```

### We can see that every class has same number of count

## Importing some necessary libraries

```python
from sklearn import tree
dec_tree = tree.DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

## Train-Test split

```python
X=df.drop("Species",axis=1)
y=df["Species"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
clf=dec_tree.fit(X_train,y_train)
clf1=dec_tree.fit(X,y)
```

## Plotting the decision tree for the training data

```python
plt.figure(figsize=(20, 20))
tree.plot_tree(clf,filled=True)
plt.show()
```

## plotting the tree for the whole data

```python
plt.figure(figsize=(20, 20))
tree.plot_tree(clf1,filled=True)
plt.show()
```

## Computing Accuracy score

```python
 y_pred=clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

## Comparing the result with random forest classifier

```python
ran_f=RandomForestClassifier()
ran_f.fit(X_train,y_train)
```

```python
y_pred1=ran_f.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```
